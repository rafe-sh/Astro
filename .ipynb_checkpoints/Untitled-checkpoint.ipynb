### Working on 0:0 Loc Dataset
(Without Limitation on stars)

```python
df = pd.read_csv("./data/1659360142355O-result.csv")
df
```

```python
df = df[~(df['parallax'].isnull()) & ~(df['bp_rp'].isnull())]
```

```python
df
```

```python
df = df[(df['phot_g_mean_flux_over_error'] > 50) & (df['phot_rp_mean_flux_over_error']>20) & (df['phot_bp_mean_flux_over_error']>20)]
```

```python
df = df[df['phot_bp_rp_excess_factor'] < (1.3 + 0.06 * ( df['phot_bp_mean_mag'] - df['phot_rp_mean_mag'])**2)]
```

```python
df = df[df['phot_bp_rp_excess_factor'] > (1.0 + 0.015 * ( df['phot_bp_mean_mag'] - df['phot_rp_mean_mag'])**2)]
```

```python
df = df[df['visibility_periods_used'] > 8]
```

```python
# exp(-0.4*(phot_g_mean_mag-19.5)

df['MAX'] = np.exp(-0.4 * (df['phot_g_mean_mag']-19.5))
```

```python
df['MAX'] = df['MAX'].apply(lambda x: 1 if(x<1) else x)
```

```python
df = df[(df['astrometric_chi2_al'] / (df['astrometric_n_good_obs_al']-5)) < 1.44 * df["MAX"]]
```

```python
# POSITIVE PARALLAX FOR LOG

df = df[df['parallax'] > 0]
df['mg'] = df['phot_g_mean_mag'] + (5 * np.log10(df['parallax']) -10)
```

```python
df
```

```python
data_plot_0_0 = df[['bp_rp', 'mg']]
```

```python
data_plot_0_0
```

```python
def using_datashader(ax, x, y):

    dsartist = dsshow(
        data_plot_0_0,
        ds.Point("bp_rp", "mg"),
        ds.count(),
#         vmin=0,
#         vmax=100,
        norm="linear",
        aspect="auto",
        ax=ax,
        cmap='hot',
        y_range=[-10,20]
    )

    plt.colorbar(dsartist)


fig, ax = plt.subplots(figsize=(10,7))

using_datashader(ax, x, y)


ax.set_xlabel('$ G _{bp} - G _{rp} $')
ax.set_ylabel('$ M_{G} $')

plt.tight_layout()
plt.show()
```
