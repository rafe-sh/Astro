### DATA with `r_est` at center

  
query:
```
SELECT source_id, r_est, r_lo, r_hi, gaia_source.source_id,gaia_source.ra,gaia_source.dec,gaia_source.parallax,gaia_source.parallax_error,gaia_source.astrometric_n_good_obs_al,gaia_source.astrometric_chi2_al,gaia_source.visibility_periods_used,gaia_source.phot_g_mean_flux_over_error,gaia_source.phot_g_mean_mag,gaia_source.phot_bp_mean_flux_over_error,gaia_source.phot_bp_mean_mag,gaia_source.phot_rp_mean_flux_over_error,gaia_source.phot_rp_mean_mag,gaia_source.phot_bp_rp_excess_factor,gaia_source.bp_rp,gaia_source.l,gaia_source.b
FROM external.gaiadr2_geometric_distance
JOIN gaiadr2.gaia_source USING (source_id)
WHERE CONTAINS(
	POINT('ICRS',gaiadr2.gaia_source.ra,gaiadr2.gaia_source.dec),
 	CIRCLE('ICRS',266.40498829,-28.93617776,1)
)=1
```

````python
full_df = pd.read_csv('../data/0_0 - 1 deg-result.csv')
````

````python
full_df.describe().transpose()
````

````python
full_df.sample(4)
````

````python
full_df = full_df[~(full_df['parallax'].isnull()) & ~(full_df['bp_rp'].isnull())]
````

````python
# POSITIVE PARALLAX FOR LOG

full_df = full_df[full_df['parallax'] > 0]
full_df['mg'] = full_df['phot_g_mean_mag'] + (5 * np.log10(full_df['parallax']) -10)
````

````python
import mpl_scatter_density 
from matplotlib.colors import LinearSegmentedColormap

hot = LinearSegmentedColormap.from_list('hot', [
    (0, '#ffffff'),
    (1e-20, '#0B0000'),
    (0.15, '#720000'),
    (0.3, '#DB0000'),
    (0.45, '#FF4500'),
    (0.6, '#FFAE00'),
    (0.75, '#FFFF24'),
    (0.9, '#FFFFAC'),
    (1, "#ffffff")
], N=65535)

def using_mpl_scatter_density(fig, df):
    ax = fig.add_subplot(1, 1, 1, projection='scatter_density')
    density = ax.scatter_density(df['bp_rp'], df['mg'], cmap=hot)
    fig.colorbar(density, label='Number of points per pixel')

    
fig = plt.figure(figsize=(9,9))
using_mpl_scatter_density(fig, full_df)

plt.ylim(20, -10)
plt.tight_layout()
plt.show()
````

````python
full_df = full_df[(full_df['phot_g_mean_flux_over_error'] > 50) & (full_df['phot_rp_mean_flux_over_error']>20) & (full_df['phot_bp_mean_flux_over_error']>20)]
````

````python
full_df = full_df[full_df['phot_bp_rp_excess_factor'] < (1.3 + 0.06 * ( full_df['phot_bp_mean_mag'] - full_df['phot_rp_mean_mag'])**2)]
````

````python
full_df = full_df[full_df['phot_bp_rp_excess_factor'] > (1.0 + 0.015 * ( full_df['phot_bp_mean_mag'] - full_df['phot_rp_mean_mag'])**2)]
````

````python
full_df = full_df[full_df['visibility_periods_used'] > 8]
````

````python
full_df['MAX'] = np.exp(-0.4 * (full_df['phot_g_mean_mag']-19.5))
````

````python
full_df['MAX'] = full_df['MAX'].apply(lambda x: 1 if(x<1) else x)
````

````python
full_df = full_df[(full_df['astrometric_chi2_al'] / (full_df['astrometric_n_good_obs_al']-5)) < 1.44 * full_df["MAX"]]
````

````python
full_df.describe().transpose()
````

````python
fig = plt.figure(figsize=(10,10))
using_mpl_scatter_density(fig, full_df)

plt.ylim(20, -10)
plt.title('0:0 dataset(filters enabled)')
plt.tight_layout()
plt.savefig('../data/pictures/0_0 dataset.jpg', dpi=200, bbox_inches='tight')
plt.show()
````
